# 3A) IMPLEMENT LOGIC GATES USING NEURAL NETWORKS

```python
class Neuron:
    def __init__(self, n_inputs, learning_rate=0.01, num_iters=1000):
        self.weights = np.random.randn(n_inputs)
        self.bias = np.random.randn()
        self.learning_rate = learning_rate
        self.num_iters = num_iters
    def activate(self, x):
        # Threshold Activation
        return 1 if x>0.5 else 0
    def fit(self, x_tr, y_tr):
        y_hat = self.activate(np.dot(x_tr, self.weights) + self.bias)
        loss = y_tr - y_hat
        self.weights += loss*self.learning_rate*x_tr
        self.bias += self.learning_rate*loss
        return loss, y_tr==y_hat
    def predict(self, x):
        return self.activate(np.dot(x, self.weights)+self.bias)
    def train(self, x_tr, y_tr):
        loss = []
        accuracy = []
        for epoch in range(self.num_iters):
            idx = random.randint(0, len(x_tr)-1)
            L, A = self.fit(x_tr[idx], y_tr[idx])
            loss.append(L)
            accuracy.append(A)
        return loss, accuracy
```

```python
random.seed(0.2323)
inputs = np.array([[random.randint(0, 1), random.randint(0, 1)] for i in range(10)])
or_outputs = np.array([i[0]|i[1] for i in inputs])
and_outputs = np.array([i[0]&i[1] for i in inputs])
nor_outputs = np.array([~(i[0]|i[1]) for i in inputs])
nand_outputs = np.array([~(i[0]&i[1]) for i in inputs])
xor_outputs = np.array([i[0]^i[1] for i in inputs])
```

# OR

```python
neuron = Neuron(2)
print(f"Perceptron Initialised With\nWeights: {neuron.weights}\nBias: {neuron.bias}\n")
loss, acc =neuron.train(inputs, or_outputs)
accuracy = 0
for i in acc:
    if i:
        accuracy+=1
accuracy /= len(acc)
print(f"After Training On OR Outputs\nWeights: {neuron.weights}\nBias: {neuron.bias}\nAccuracy: {accuracy*100:.2f}%")
print(f"Input: {[0,1]} Expected Output: 1 Predicted = {neuron.predict(np.array([0,1]))}")
plt.plot(range(len(loss)), loss)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Epoch VS Loss")
```

# 3b) MULTI LAYERED PERCEPTRON

```python
import numpy as np

class SimpleMLP:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.01):
        # Initialize weights and biases
        self.weights1 = np.random.randn(input_size, hidden_size) * 0.01  # Input to hidden layer
        self.bias1 = np.zeros((1, hidden_size))
        self.weights2 = np.random.randn(hidden_size, output_size) * 0.01  # Hidden to output layer
        self.bias2 = np.zeros((1, output_size))
        self.learning_rate = learning_rate
        self.loss = []
    def relu(self, x):
        return np.maximum(0, x)

    def relu_derivative(self, x):
        return (x > 0).astype(float)

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))  # Stability trick
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def forward(self, X):
        # Forward pass
        self.z1 = np.dot(X, self.weights1) + self.bias1
        self.a1 = self.relu(self.z1)  # Activation for hidden layer
        self.z2 = np.dot(self.a1, self.weights2) + self.bias2
        self.a2 = self.softmax(self.z2)  # Output probabilities
        return self.a2

    def backward(self, X, y, y_pred):
        # Compute gradients
        m = X.shape[0]
        dz2 = y_pred - y  # Gradient for output layer
        dw2 = np.dot(self.a1.T, dz2) / m
        db2 = np.sum(dz2, axis=0, keepdims=True) / m

        dz1 = np.dot(dz2, self.weights2.T) * self.relu_derivative(self.z1)  # Gradient for hidden layer
        dw1 = np.dot(X.T, dz1) / m
        db1 = np.sum(dz1, axis=0, keepdims=True) / m

        # Update weights and biases
        self.weights2 -= self.learning_rate * dw2
        self.bias2 -= self.learning_rate * db2
        self.weights1 -= self.learning_rate * dw1
        self.bias1 -= self.learning_rate * db1

    def train(self, X, y, epochs=1000):
        for epoch in range(epochs):
            # Forward pass
            y_pred = self.forward(X)                
            # Backward pass
            self.backward(X, y, y_pred)
            loss = -np.mean(np.sum(y * np.log(y_pred + 1e-8), axis=1))  # Cross-entropy loss
            self.loss.append(loss)
            # Print loss occasionally
            if epoch % 100 == 0:
                loss = -np.mean(np.sum(y * np.log(y_pred + 1e-8), axis=1))  # Cross-entropy loss
                print(f"Epoch {epoch}, Loss: {loss}")
        

    def predict(self, X):
        # Predict class labels
        y_pred = self.forward(X)
        return np.argmax(y_pred, axis=1)


```
